##                       Activity 2: Data Cleaning and Feature Engineering

```python
#. Duplicates
df.drop_duplicates(inplace=True)
```

Fill the missing values as per the strategy of the learners, one example is given here  

```python
df['continent'].fillna('Unknown', inplace=True)
```

```python
df['total_cases'].fillna(df['total_cases'].median(), inplace=True)
df['new_cases'].fillna(0, inplace=True)
df['total_deaths'].fillna(df['total_deaths'].median(), inplace=True)
df['new_deaths'].fillna(0, inplace=True)
```

# Plotting New Cases Over Time

```python
import matplotlib.pyplot as plt


# Plotting using Matplotlib
plt.figure(figsize=(15, 8))
plt.plot(us_cases['date'], us_cases['new_cases'], label='New Cases')
plt.title('COVID-19 New Cases in the United States')
plt.xlabel('Date')
plt.ylabel('New Cases')
plt.legend()
plt.grid(True)
plt.show()
```

# Smoothen out the line with a 7 day moving average

#### Some laboratories do not report data on weekends, so thats why we have such a volatile chart with huge swings up and down

#### By using the 7 day moving average, which is standard when reporting new covid cases, this allows us to get a clearer picture of the cases over time

```python

# Calculate the 7-day rolling average
us_cases['7_day_avg'] = us_cases['new_cases'].rolling(window=7).mean()

# Plotting using Pandas and Matplotlib
plt.figure(figsize=(15, 8))
us_cases.plot(x='date', y='7_day_avg', figsize=(15, 8), label='7-Day Rolling Average')
plt.title('COVID-19 7-Day Rolling Average in the United States')
plt.xlabel('Date')
plt.ylabel('7-Day Rolling Average')
plt.legend()
plt.grid(True)
plt.show()
```

# View only the new cases after May 2021


Remember that `.loc` allows us to select rows in a DataFrame

`us_cases['date'] > '2021-05-01'` allows us to access only the rows where the
date is past May 2021

```python
afterMay2021 = us_cases.loc[us_cases['date'] > '2021-05-01']

afterMay2021['7_day_avg'] = afterMay2021['new_cases'].rolling(window=7).mean()
```

```python
# Plotting using Matplotlib
plt.figure(figsize=(15, 8))
plt.plot(afterMay2021['date'], afterMay2021['7_day_avg'], label='7-Day Rolling Average')
plt.title('COVID-19 7-Day Rolling Average (After May 2021)')
plt.xlabel('Date')
plt.ylabel('7-Day Rolling Average')
plt.legend()
plt.grid(True)
plt.show()
```

Daily Average New Cases and Deaths:
Calculate and visualize the daily average new cases and deaths globally.

```python
df['date'] = pd.to_datetime(df['date'])
df['day_of_week'] = df['date'].dt.day_name()

# Calculate daily average new cases and deaths
daily_avg_cases = df.groupby('day_of_week')['new_cases'].mean()
daily_avg_deaths = df.groupby('day_of_week')['new_deaths'].mean()

# Plotting
plt.figure(figsize=(12, 6))
sns.barplot(x=daily_avg_cases.index, y=daily_avg_cases, label='New Cases')
sns.barplot(x=daily_avg_deaths.index, y=daily_avg_deaths, label='New Deaths')
plt.title('Daily Average New Cases and Deaths Globally')
plt.xlabel('Day of Week')
plt.ylabel('Average Count')
plt.legend()
plt.show()
```
